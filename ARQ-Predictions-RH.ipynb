```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from fbprophet import Prophet
from sklearn.metrics import r2_score,mean_squared_error
from statsmodels.tsa.arima_model import ARIMA
from sklearn import linear_model as lm
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
from scipy.interpolate import UnivariateSpline
data=pd.read_pickle('AQI-FE1-Binarized.pkl')
dx=data[[ 'Month', 'Quarter', 'Dayofweek', 'HourofDay',
       'DayofMonth', 'Dayoftheyear', 'Monday', 'Tuesday', 'Wednesday',
       'Thursday', 'Friday', 'Saturday', 'Sunday','RH']].copy(deep=True)
dx.sort_index(inplace=True)
```

```python
fig, ax = plt.subplots()
#finding trend using rolling window operation
dx1=dx['RH'].rolling(window=72, center=True, min_periods=24).mean()
dx.plot(y='RH',ax=ax)
dx1.plot(y='RH', ax=ax)
```

```python
fig, ax = plt.subplots()
#finding trend using rolling window operation
dx2=dx['RH'].rolling(window=48, center=True, min_periods=24).mean()
dx.plot(y='RH',ax=ax)
dx2.plot(y='RH', ax=ax)
```

```python
fig, ax = plt.subplots()
#finding trend using rolling window operation
dx3=dx['RH'].rolling(window=24, center=True, min_periods=24).mean()
dx.plot(y='RH',ax=ax)
dx3.plot(y='RH', ax=ax)
```

```python
fig, ax = plt.subplots()
#finding trend using rolling window operation
dx4=dx['RH'].rolling(window=12, center=True, min_periods=12).mean()
dx.plot(y='RH',ax=ax)
dx4.plot(y='RH', ax=ax)
```

```python
dx['rolling-72']=dx['RH'].rolling(window=72, center=True, min_periods=6).mean()
dx['rolling-48']=dx['RH'].rolling(window=48, center=True, min_periods=6).mean()
dx['rolling-24']=dx['RH'].rolling(window=24, center=True, min_periods=6).mean()
dx['rolling-12']=dx['RH'].rolling(window=12, center=True, min_periods=6).mean()
```

```python
train=dx.loc[:'2005-3-1'].copy(deep=True)
test=dx.loc['2005-3-2':].copy(deep=True)
```

```python
attb=['rolling-72','rolling-48','rolling-24','rolling-12']
RM = lm.LinearRegression()
RM.fit(train[attb],train['RH'])
```

```python
pred = RM.predict(test[attb])
pred=pd.DataFrame(pred,columns=['Prediction'])
pred.index=test.index
test1=pd.concat([test,pred],axis=1)
```

```python
fig, ax = plt.subplots()
#finding trend using rolling window operation
dx.plot(y='RH',ax=ax)
test1.plot(y='Prediction',ax=ax)
dx.plot(y='RH')
```

```python
mean_squared_error(test1['RH'], test1['Prediction'])
```

```python
r2_score(test1['RH'], test1['Prediction'])
```

```python
reg2 = RandomForestRegressor(random_state=1, n_estimators=100,max_depth=30,min_samples_leaf=3,min_samples_split=5)
reg2.fit(train[attb],train['RH'])
pred = reg2.predict(test[attb])
pred=pd.DataFrame(pred,columns=['Prediction'])
pred.index=test.index
test1=pd.concat([test,pred],axis=1)
fig, ax = plt.subplots()
#finding trend using rolling window operation
dx.plot(y='RH',ax=ax)
test1.plot(y='Prediction',ax=ax)
dx.plot(y='RH')
```

```python
mean_squared_error(test1['RH'], test1['Prediction'])
```
