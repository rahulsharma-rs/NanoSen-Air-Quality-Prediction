```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('grid-load.csv')
url='http://www.ghanagrid.com/index.php/loadprofile?disptime=1575590400'
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['Date']=pd.to_datetime(data['Date'])
data['Time']=pd.to_timedelta(data["Time"],unit='s')
data.index=data["Date"] + data["Time"]
```

```python
data.index
```

```python
data.plot(y='Load')
plt.show()
```

```python
fig, ax = plt.subplots()
#finding trend using rolling window operation
data['r1']=data['Load'].rolling(window=48, center=True, min_periods=2).mean()
data.plot(y='Load',ax=ax)
data.plot(y='r1', ax=ax)
```
